```
import numpy as np
import datetime
import pandas as pd
!pip install tensorflow
import tensorflow as tf

from tensorflow.keras.datasets import fashion_mnist
```

<h1><center><b></b>Data Preprocessing</center></h1>

```
#loading dataset
(X_train,y_train),(X_test,y_test)=fashion_mnist.load_data()
```

```
#normalizing the images 
# by dividing with max available value it will be converted inot pixels range of [0,1]
X_train=X_train/255.0
X_test=X_test/255.0
```

```
X_train.shape
```

```
#reshaping the vector 1x1
X_train=X_train.reshape(-1,28*28)
X_train.shape
```

```
X_test=X_test.reshape(-1,28*28)
X_test.shape
```

```
#Artificial Neural Network
#simple Sequential model
model=tf.keras.models.Sequential()
```

```
#Adding a fullly connected hidden layer
#layers=128
#shape=(784,)
#activation function=ReLU
model.add(tf.keras.layers.Dense(units=128,activation='relu',input_shape=(784,)))
# model.add(tf.keras.layers.Dense(units=128,activation='relu',input_shape=(784,)))
```

```
#adding second layer with dropout

#Dropout is a regularisation technique when we randomnly set neurons to 0. 
#this way neurons dont get updated while trainin.

model.add(tf.keras.layers.Dropout(0.2))
```

```
#adding output layer
model.add(tf.keras.layers.Dense(units=10,activation='softmax'))
```

```
#Compiling model
#optimizer=Adam
#loss= Sparse Softmax(categorical crossentropy)
model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['sparse_categorical_accuracy'])
```

```
model.summary()
```

```
model.fit(X_train,y_train,epochs=15)
```

```
#model_evaluation
test_loss,test_accruacy=model.evaluate(X_test,y_test)
test_accruacy
```
